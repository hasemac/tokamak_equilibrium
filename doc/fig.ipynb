```python
import plotly.graph_objects as go
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
y0 = -x**2 + 1
y1 = -x**6 + 1
y2 = -x**50 + 1
```

```python
dat = [
    go.Scatter(x=x, y=y0, name='parabolic'), 
    go.Scatter(x=x, y=y1, name='6th degree'),
    go.Scatter(x=x, y=y2, name='50th degree'),
]
lay = go.Layout(
    width = 400, height = 250,
    margin= {'l':20, 'r':20, 't':20, 'b':20},
    xaxis = {'title':'Normalized minor radius', 'range':[0, 1.1]},
    yaxis = {'title':'Normalized Jt', },
    legend= {'x':0.01, 'y': 0.01},
)
fig = go.Figure(data=dat, layout=lay)

fig.show()
```
